```python
# Pulling together film and category ID
# Query ask for Title, Length (runtime), and film category
# Goal was to compare runtime and film category. In 'film database', category is coded. Needed to use codes in category_id to
#compare properly. Had to bring in film_category database to connect category id to film.
stmt = """SELECT film.title, film.length, category.name AS Category FROM film 
        JOIN film_category ON film.film_id = film_category.film_id
        JOIN category on film_category.category_id = category.category_id
        ;
"""
results_join = connection.execute(stmt).fetchall()
```

```python
len(results_join)
```

```python
df = pd.DataFrame(results_join)
df.columns = results_join[0].keys()
```

```python
# Average runtime for each category
df.groupby('Category')['length'].apply(np.mean).sort_values(ascending = False)
```
